```python
# plot graph based on Sorted Price

df01 = df.groupby('Airline')['Price'].sum()
df01_sorted = df01.sort_values(ascending=False)
df01 = pd.DataFrame({'Airline': df01_sorted.index, 'Price': df01_sorted.values})
df01
```

```python

sns.barplot(x=df01['Airline'],y=df01['Price'],palette='Accent',errorbar=None)
plt.xticks(rotation=45)
plt.show()
```

```python
# Add Hue of Source Multivariate
import seaborn as sns

plt.figure(figsize=(12,3))

sns.barplot(x=df['Airline'] ,y=df['Price'],hue=df['Source'],errorbar=None)
plt.xticks(rotation=45,fontsize=14)
plt.yticks(fontsize=14)
plt.title("Airline with Source Vs Price",fontsize=20)
plt.show()
```
