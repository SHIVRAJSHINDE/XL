```python

def replaceWithMode(columnName):
    for i in columnName:
        modeValue = df[i].mode()[0]
        # print(modeValue,i)
        df[i].fillna(modeValue,inplace=True)
    return df 

# Replacing the 'Self_Employed','Married' with mode  
columnName = replaceMode

df = replaceWithMode(columnName)

df
```

```python

def replaceWithMean(columnName1):
    for i in columnName1:
        meanOfColumn = df[i].mean()
        df[i].fillna(meanOfColumn,inplace=True)
    return df 

columnName1 = ['LoanAmount']
df = replaceWithMean(columnName1)

df
```
