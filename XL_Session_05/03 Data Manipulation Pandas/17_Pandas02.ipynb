# <h2 style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-weight: bold; text-align: center; padding-bottom: 10px; border-bottom: 2px solid #264653;">Handelling Missing Value </h2>

<p style="color: #264653; font-family: 'Candara', sans-serif; font-size: 18px; line-height: 1.8; margin-top: 20px;">
    Handling missing values is essential in data cleaning. Pandas provides two primary methods for dealing with missing data: <b>fillna</b> and <b>dropna</b>.
</p>

<ol style="font-family: 'Candara', sans-serif; font-size: 18px; line-height: 1.8; padding-left: 20px; margin-top: 10px; list-style-type: none;">
    <li style="color: #2a9d8f;"><b>.fillna</b>
        <p style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-size: 16px; margin: 0;">
            <i>Description:</i> The <code>fillna</code> method is used to fill missing values in a DataFrame or Series with a specified value or method.<br>
            <i>Syntax:</i> <code>df.fillna(value=None, method=None, axis=None, inplace=False)</code><br>
            <i>Parameters:</i>
            <ul>
                <li><b>value</b>: Scalar, dict, Series, or DataFrame. The value to use for filling missing data.</li>
                <li><b>method</b>: {'backfill', 'bfill', 'pad', 'ffill', None}, default None. Method for filling missing values.</li>
                <li><b>axis</b>: {0 or 'index', 1 or 'columns'}, default None. The axis along which to fill missing values.</li>
                <li><b>inplace</b>: bool, default False. Whether to modify the DataFrame in place or return a new DataFrame.</li>
            </ul>
            <i>Examples:</i>
            <ul>
                <li><code>df.fillna(0)</code> - Replaces all NaN values with 0.</li>
                <li><code>df.fillna(method='ffill')</code> - Forward fills missing values using the last valid observation.</li>
                <li><code>df.fillna(method='bfill')</code> - Backward fills missing values using the next valid observation.</li>
                <li><code>df.fillna({'col1': 0, 'col2': 1})</code> - Fills NaNs in 'col1' with 0 and in 'col2' with 1.</li>
            </ul>
        </p>
    </li>
    <li style="color: #2a9d8f;"><b>.dropna</b>
        <p style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-size: 16px; margin: 0;">
            <i>Description:</i> The <code>dropna</code> method is used to remove missing values from a DataFrame or Series.<br>
            <i>Syntax:</i> <code>df.dropna(axis=0, how='any', subset=None, inplace=False)</code><br>
            <i>Parameters:</i>
            <ul>
                <li><b>axis</b>: {0 or 'index', 1 or 'columns'}, default 0. The axis along which to drop missing values.</li>
                <li><b>how</b>: {'any', 'all'}, default 'any'. Drop rows or columns based on the presence of missing values.</li>
                <li><b>subset</b>: column label or sequence of labels, optional. Only consider certain columns for missing values.</li>
                <li><b>inplace</b>: bool, default False. Whether to modify the DataFrame in place or return a new DataFrame.</li>
            </ul>
            <i>Examples:</i>
            <ul>
                <li><code>df.dropna()</code> - Drops all rows where any value is NaN.</li>
                <li><code>df.dropna(axis=1)</code> - Drops all columns where any value is NaN.</li>
                <li><code>df.dropna(how='all')</code> - Drops rows where all values are NaN.</li>
                <li><code>df.dropna(subset=['col1', 'col2'])</code> - Drops rows where NaNs are present in 'col1' or 'col2'.</li>
            </ul>
        </p>
    </li>
</ol>

# <h2 style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-weight: bold; text-align: center; padding-bottom: 10px; border-bottom: 2px solid #264653;">Adding New Columns to a DataFrame </h2>


<p style="color: #264653; font-family: 'Candara', sans-serif; font-size: 18px; line-height: 1.8; margin-top: 20px;">
    Adding new columns to a DataFrame can be done in several ways. Below are the most popular methods:
</p>

<ol style="font-family: 'Candara', sans-serif; font-size: 18px; line-height: 1.8; padding-left: 20px; margin-top: 10px; list-style-type: none;">
    <li style="color: #2a9d8f;"><b>Directly Adding a Column</b>
        <p style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-size: 16px; margin: 0;">
            <i>Description:</i> You can directly add a new column to a DataFrame by assigning values to a new column label.<br>
            <i>Syntax:</i> <code>df['new_column_name'] = value</code><br>
            <i>Example:</i> <code>df['NewColumn'] = [1, 2, 3, 4]</code>
        </p>
    </li>
    <li style="color: #2a9d8f;"><b>Adding a Column Based on Existing Column</b>
        <p style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-size: 16px; margin: 0;">

<ol style="font-family: 'Candara', sans-serif; font-size: 18px; line-height: 1.8; padding-left: 20px; margin-top: 10px; list-style-type: none;">

        <p style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-size: 16px; margin: 0;">

                <li><b>overwrite</b>: bool, default True. Whether to overwrite existing values with the values from <code>other</code>.</li>
                <li><b>filter_func</b>: function, default None. Function to filter the DataFrame based on specific conditions.</li>

            <i>Examples:</i>
            <ul>
                <li><code>df.update(df2)</code> - Updates <code>df</code> with non-NA values from <code>df2</code>.</li>

```python
import pandas as pd

df1 = pd.DataFrame(
        {"A": ["A0", "A1", "A2", "A3"],
            "B": ["B0", "B1", "B2", "B3"],
        },
        index=[0, 1, 2, 3],
    )
    

df2 = pd.DataFrame(
        {   "A": ["A0", "A1", "A2", "A3"],
            "B": ["B4", "B5", "B6", "B7"],
                 },
        index=[4, 5, 6, 7],
    )
    

df3 = pd.DataFrame({"A": ["A0", "A1", "A2", "A3"],
                    "B": ["B8", "B9", "B10", "B11"],},
                    index=[8, 9, 10, 11],)

print(df1)
print()
print(df2)
print()
print(df3)
```

```python
'''Join Dataframe on top of the above '''
allDf = [df1,df2,df3]
# row = 0
# column = 1
pd.concat(allDf,axis=0)
```

```python
'''Join Dataframe on top of the above '''
allDf = [df1,df2,df3]
# row = 0
# column = 1
pd.concat(allDf,axis=1)
```

```python
'''Join Dataframe Side by Side '''
# here common key is not considered
allDf = [df1, df2, df3]

result_df = pd.concat(allDf, axis=1)
result_df
```

<h2 style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-weight: bold; text-align: center; padding-bottom: 10px; border-bottom: 2px solid #264653;">
    Pandas GroupBy
</h2>

<p style="color: #264653; font-family: 'Candara', sans-serif; font-size: 18px; line-height: 1.8; margin-top: 20px;">
    The <code>groupby</code> method in Pandas is used to group data based on one or more columns and perform operations on these groups. This is useful for aggregating and summarizing data.
</p>

<h3 style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-size: 20px; margin-top: 20px;">
    Syntax:
</h3>
<p style="color: #264653; font-family: 'Candara', sans-serif; font-size: 18px; line-height: 1.8; margin-top: 10px;">
    <code>df.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, dropna=True)</code>
</p>

<h3 style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-size: 20px; margin-top: 20px;">
    Parameters:
</h3>
<ul style="font-family: 'Candara', sans-serif; font-size: 18px; line-height: 1.8; padding-left: 20px; margin-top: 10px; list-style-type: square;">
    <li><b>by:</b> The column(s) to group by. It can be a column name, list of column names, or a function.</li>
    <li><b>axis:</b> The axis to group along (default is 0 for rows).</li>
    <li><b>level:</b> If the axis is a MultiIndex (hierarchical), group by a particular level.</li>
    <li><b>as_index:</b> Whether to set the group labels as the index (default is True).</li>
    <li><b>sort:</b> Whether to sort the group labels (default is True).</li>
    <li><b>group_keys:</b> Whether to include group keys in the result (default is True).</li>
    <li><b>dropna:</b> Whether to drop groups with missing values (default is True).</li>
</ul>

<h3 style="color: #2a9d8f; font-family: 'Candara', sans-serif; font-size: 20px; margin-top: 20px;">

        <code>df.groupby(['column1', 'column2']).sum()</code> - Groups by multiple columns and calculates the sum of each group.</li>
